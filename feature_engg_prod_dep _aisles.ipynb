```python
import pandas as pd
import numpy as np
```

```python
df_dep = pd.read_csv("departmentseda.csv")
df_al  = pd.read_csv("aisleseeda.csv")
df_pd  = pd.read_csv("productseda.csv")
```

```python
products_enriched = (
    df_pd
    .merge(df_al, on="aisle_id", how="left")
    .merge(df_dep, on="department_id", how="left")
)

len(products_enriched) == len(df_pd)
```
